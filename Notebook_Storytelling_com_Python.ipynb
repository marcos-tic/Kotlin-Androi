# Storytelling com Python
- É a capacidade de transmitir conteúdo (contar história) por meio de enredo elaborado e de narrativa envolvente, usando palavras e recursos audiovisuais;
- O O objetivo é promover o seu negócio, vender seus serviços ou se comunicar com caráter persuasivo.

- Nesta aula, vamos construir visualizações gráficas usando matplotlib, fornecendo detalhes para uma boa comunicação visual gráfica

### Adicionando elementos

```python
# Criar uma figura para grafico de barras
fig, ax = plt.subplots(figsize=(8, 4))

# Plotar antes e depois saparadamente
width_bar = 0.40
barras_antes = ax.bar(x-width_bar/2, df.Antes, width_bar, color=color_antes)
barras_depois = ax.bar(x+width_bar/2, df.Depois, width_bar, color=color_depois)


##### REMOVER ELEMENTOS NÃO ESSENCIAIS DA FIGURA

# Remover grids e eixos
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# Remover eixo y
plt.gca().axes.get_yaxis().set_visible(False)


##### ADICIONAR ELEMENTOS DE TEXTOS

# labels do eixo
ax.set_xticks(x)
ax.set_xticklabels(labels);

x_text_antes = 0.2

# Adicionando um título
plt.text(x=x_text_antes, 
         y=52, 
         s='Como você se sente em relação à Ciência',
         fontsize=16, 
         color=cinza_escuro,
         weight='bold');

# TEXTO: Resultados antes da campanha
plt.text(x_text_antes, 
         y=49,
         s="$\\bf{Antes}$ do programa\n"
         "a maioria das crianças\n"
         "se sentia BEM em\n"
         "relação à ciência.",
         verticalalignment="top",
         fontsize=12,
         color=cinza_medio)

# # TEXTO: Resultados da campanha
plt.text(x=5.5, 
         y= 39,
         s="$\\bf{DEPOIS}$\n"
         "do programa\n"
         "mais crianças tinham\n"
         "algum interesse e\n"
         "muito interesse em ciência.",
         verticalalignment="top",
         fontsize=12,
         color=azul_escuro);
```

### Adicionar porcentagem em cada barra

```python
#criar função para dar labels automaticamente
def set_label_y_in_bar(bar):
    for b in bar:
        height_bar = b.get_height()
        ax.annotate(
                    '{}%'.format(height_bar),
                    xy=(b.get_x() + b.get_width()/2,  height_bar),
                    xytext=(0, -20),  # offset
                    textcoords="offset points",
                    ha='center',
                    va='bottom',
                    fontsize=10,
                    color="white",
                    )
```

```python
# Criar uma figura para grafico de barras
fig, ax = plt.subplots(figsize=(10, 5))

# Plotar antes e depois saparadamente
width_bar = 0.40
barras_antes = ax.bar(x-width_bar/2, df.Antes, width_bar, color=color_antes)
barras_depois = ax.bar(x+width_bar/2, df.Depois, width_bar, color=color_depois)


##### REMOVER ELEMENTOS NÃO ESSENCIAIS DA FIGURA

# Remover grids e eixos
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# Remover eixo y
plt.gca().axes.get_yaxis().set_visible(False)


##### ADICIONAR ELEMENTOS DE TEXTOS

# labels do eixo
ax.set_xticks(x)
ax.set_xticklabels(labels)


x_text_antes = 0.2

# Adicionando um título
plt.text(x=x_text_antes, 
         y=52, 
         s='Como você se sente em relação à Ciência',
         fontsize=16, 
         color=cinza_escuro,
         weight='bold');

# TEXTO: Resultados antes da campanha
plt.text(x_text_antes, 
         y=49,
         s="$\\bf{Antes}$ do programa\n"
         "a maioria das crianças\n"
         "se sentia BEM em\n"
         "relação à ciência.",
         verticalalignment="top",
         fontsize=12,
         color=cinza_medio)

# TEXTO: Resultados da campanha
plt.text(x=5.5, 
         y= 39,
         s="$\\bf{DEPOIS}$\n"
         "do programa\n"
         "mais crianças tinham\n"
         "algum interesse e\n"
         "muito interesse em ciência.",
         verticalalignment="top",
         fontsize=12,
         color=azul_escuro);


set_label_y_in_bar(barras_antes)
set_label_y_in_bar(barras_depois)

# plotar o gráfico
fig.tight_layout()
plt.show()
```
